```python
#Imports pandas
import pandas as pd

#Reads the CSV
df = pd.read_csv("titanic.csv")

#Prints the first 5 lines
print("Prints first 5 lines: ")
print(df.head())
```

```python
#Combination of columns and data
print("Selecting and 5 data: ")
print(df["Cabin"].head)
print("Show only data less than 2: ")
print(df[df["Pclass"]<2])
print("Combine data Cabin and Embarked: ")
print(df.loc[df["Cabin"].notna(), ["Cabin", "Embarked", "Sex"]])
print(df.loc[df["Cabin"]=="C85",["Embarked","Sex"]])

print("Null Values: ")
Nullvalues = df.isnull().sum()
print(Nullvalues)

#Check the data and describe
print("Stadistical summary of the data: ")
print(df.describe())
```
